```python
import numpy as np
import cv2
from tensorflow.keras.models import load_model

# Function to load and preprocess image
def load_and_preprocess_image(img_path):
    print(f"Loading image: {img_path}")  # Debugging
    img = cv2.imread(img_path)
    
    if img is None:
        print(f"Error loading image: {img_path}")
        return None  # Return None if image loading fails

    img = cv2.resize(img, (128, 128))  # Resize image to match model input size
    img = img.astype("float32") / 255.0  # Normalize image
    img_array = img.reshape((1, 128, 128, 3))  # Reshape to match model's input format
    print(f"Image shape after preprocessing: {img_array.shape}")  # Debugging
    return img_array

# Load the trained model
model_path = r'C:\Users\alank\OneDrive\Desktop\deepfake detection\best_deepfake_model.keras'
best_model = load_model(model_path)

# Test image path (modify with your own image path)
test_image_path = r'C:\Users\alank\OneDrive\Desktop\deepfake detection\Screenshot 2024-03-21 132553.png'  # Modify with the path to your test image

# Load and preprocess the image
img_array = load_and_preprocess_image(test_image_path)

if img_array is not None:
    # Make prediction
    prediction = best_model.predict(img_array)[0][0]  # Get prediction (output of model)
    
    # Print prediction result
    result = 'Real' if prediction >= 0.5 else 'Fake'
    print(f"Prediction: {result} (Prediction probability: {prediction})")
else:
    print("Failed to process the image.")
```
